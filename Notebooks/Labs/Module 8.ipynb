1. The Person.Person table contains names for several tables. Write the following queries and then use UNION ALL operator to return one result set. Each query should return the ID, role (a literal string), first name and last name.

```sql
USE AdventureWorks2019
GO

SELECT emp.BusinessEntityID, p.FirstName, p.LastName, 'Employee' AS Role
FROM Person.Person as p
INNER JOIN HumanResources.Employee as emp
ON emp.BusinessEntityID = p.BusinessEntityID
UNION ALL

SELECT jc.BusinessEntityID, p.FirstName, p.LastName, 'Job Candidate' as Role
FROM Person.Person as p
INNER JOIN HumanResources.JobCandidate AS jc
ON jc.BusinessEntityID = p.BusinessEntityID
UNION ALL

SELECT c.PersonID, p.FirstName, p.LastName, 'Customer' as Role
FROM Person.Person as p
INNER JOIN Sales.Customer as c
ON c.PersonID = p.BusinessEntityID
UNION ALL

SELECT sp.BusinessEntityID, p.FirstName, p.LastName, 'Salesperson' as Role
FROM Person.Person as p
INNER JOIN Sales.SalesPerson as sp
ON sp.BusinessEntityID = p.BusinessEntityID
```
